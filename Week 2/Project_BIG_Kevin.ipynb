```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine.data import diamonds
import warnings
warnings.filterwarnings("ignore")
diamonds
```

```python
# cek deskripsi data
diamonds.describe()
```

```python
# cek nilai 0 pada x, y, z
# variabel x, y, z adalah ukuran diamonds/berlian, jadi tidak mungkin 0
diamonds.loc[(diamonds['x']==0) | (diamonds['y']==0) | (diamonds['z']==0)]
```

```python
# cek jumlah pastinya nilai 0 pada x, y, z
len(diamonds[(diamonds['x']==0) | (diamonds['y']==0) | (diamonds['z']==0)])
```

```python
# hapus nilai 0
diamonds = diamonds[(diamonds[['x','y','z']] != 0).all(axis=1)]
```

```python
# cek kembali
len(diamonds[(diamonds['x']==0) | (diamonds['y']==0) | (diamonds['z']==0)])
```

```python
data_corr = diamonds.corr()
sns.heatmap(data=data_corr, square=True , annot=True, cbar=True)
```

```python
plt.title('Harga Berlian Berdasarkan Karat')
sns.set(style="whitegrid") 
sns.lineplot( diamonds['carat'], diamonds['price'], ci=None)
# carat mempengaruhi harga, semakin besar karat semakin besar harga
# berat = carat
```

**Pengaruh Carat Terhadap Price**
>semakin besar carat berliannya maka harga semakin tinggi

```python
plt.title('Harga Berlian Berdasarkan Kedalaman')
sns.set(style="whitegrid") 
sns.lineplot( diamonds['depth'], diamonds['price'], ci=None)
# depth mempengaruhi harga, semakin besar depth semakin kecil harga atau murah
```

**Pengaruh Depth Terhadap Price**
>Semakin besar kedalaman berliannya maka semakin murah

```python
plt.title('Harga Berlian Berdasarkan Kejernian')
sns.set(style="whitegrid") 
sns.barplot( diamonds['clarity'], diamonds['price'], ci=None)
# 
```

**Pengaruh Clarity terhadap price**
>rata-rata clarity yang bagus murah harganya

```python
plt.title('Harga Berlian Berdasarkan Warna')
sns.set(style="whitegrid") 
sns.barplot( diamonds['color'], diamonds['price'], ci=None)
```

**Pengaruh Color Terhadap Price**
>ternyata semakin buruk warna berliannya maka semakin mahal harganya

```python
plt.title('Harga Berlian Berdasarkan Ukiran')
sns.set(style="whitegrid") 
sns.barplot( diamonds['cut'], diamonds['price'], ci=None)
```

**Pengaruh Cut Terhadap Price**
>ideal sebagai bentuk berlian yang paling bagus murah harganya
