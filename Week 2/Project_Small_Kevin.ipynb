```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from plotnine.data import msleep
from plotnine import ggplot, aes, geom_point
msleep
```

```python
modus = msleep['vore'].mode()[0]
modus2 = msleep['conservation'].mode()[0]
modus3 = msleep['sleep_rem'].mode()[0]
modus4 = msleep['sleep_cycle'].mode()[0]
modus5 = msleep['brainwt'].mode()[0]

msleep['vore'] = msleep['vore'].fillna(modus)
msleep['conservation'] = msleep['conservation'].fillna(modus2)
msleep['sleep_rem'] = msleep['sleep_rem'].fillna(modus3)
msleep['sleep_cycle'] = msleep['sleep_cycle'].fillna(modus4)
msleep['brainwt'] = msleep['brainwt'].fillna(modus5)
```

```python
sns.set(style="whitegrid") 
plt.title('Jumlah waktu Bangun')
sns.scatterplot( msleep['sleep_total'], msleep['awake'],ci=None)
```

**Pengaruh Waktu Tidur Terhadap Waktu Bangun**
>Semakin banyak waktu tidur semakin cepat waktu bangun

```python
sns.scatterplot( msleep['sleep_total'], msleep['sleep_rem'],ci=None)
```

**Pengaruh Waktu Tidur Terhadap Waktu Mimpi**
>semakin banyak waktu tidur semakin banyak waktu bermimpi

```python
sns.set(style="whitegrid") 
plt.title('Pengaruh Berat Badan dengan Berat Otak')
sns.lineplot( msleep['bodywt'], msleep['brainwt'],ci=None)
```

**Pengaruh Berat Badan Terhadap Berat Otak**
>Semakin besar berat badan maka semakin besar ukuran otak

```python
sns.set(style="whitegrid") 
plt.title('Pengaruh Berat Otak dengan Lamanya Waktu Tidur')
sns.lineplot( msleep['sleep_total'], msleep['brainwt'],ci=None)
```

**Pengaruh jumlah waktu tidur terhadap ukuran otak**
>Semakin sedikit waktu tidur maka semakin besar ukuran otak
